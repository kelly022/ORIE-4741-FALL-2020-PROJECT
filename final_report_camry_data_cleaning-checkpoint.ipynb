# One hoting the trim level

```python
import pandas as pd
sheet1=pd.read_excel('Camry.xlsx',sheet_name='L')
sheet2=pd.read_excel('Camry.xlsx',sheet_name='LE')
sheet3=pd.read_excel('Camry.xlsx',sheet_name='SE')
sheet4=pd.read_excel('Camry.xlsx',sheet_name='XLE')
sheet5=pd.read_excel('Camry.xlsx',sheet_name='XSE')
sheet_list=[sheet1,sheet2,sheet3,sheet4,sheet5]
trim_level_standard=['L','LE','SE','XLE','XSE']
for x in range(5):
    for y in range(5):
        if y==x:
            sheet_list[x][trim_level_standard[y]]=1
        else:
            sheet_list[x][trim_level_standard[y]]=0
total_sheet=pd.concat([sheet1,sheet2,sheet3,sheet4,sheet5])
a=total_sheet
```

# Split the make and model

```python
make_first=[]
make_second=[]
index=[]
for x in range(len(list(a['make_and_model']))):
    k=list(a['make_and_model'])[x].split()
    if len(k)==2:
        make_first.append(k[0])
        make_second.append(k[1])
    elif k[0]=='BMW':
        make_first.append(k[0])
        make_second.append(k[1]+' '+k[2])
    else:
        index.append(x)
#a["make_first"]=make_first
#a['make_second']=make_second
```

```python
b=a[~a.index.isin(index)]
```

```python
b['make_first']=make_first
b['make_second']=make_second
```
