# i) Further cleaning the data, change accident, past owners to int and one-hot encoding the usage

```python
import pandas as pd
a=pd.read_csv('data_clean.csv')
```

```python
a=a.drop(columns=['Unnamed: 0','Unnamed: 0.1'])
```

```python
a=a.drop(columns=['to be processed'])
```

Further cleaning the data

```python
k=a['Accident']
a['Accident']=k.apply(lambda x: 0 if x.split()[0]=="No" else x.split()[0])
```

```python
k=a['Past Owners']
a['Past Owners']=k.apply(lambda x: x.split()[0])
```

Change to use to one-hot coding

```python
a.loc[:,'is_Fleet_Use']=pd.get_dummies(a['Usage'])[' Fleet use']
a.loc[:,'is_Personal_Use']=pd.get_dummies(a['Usage'])[' Personal use']
```

# j) now perform linear regression 
use the mileage, year, useage, accident and past owners as feature and listed_price as output to fit in linear regression model

```python
from sklearn.linear_model import LinearRegression

outcome_name = 'listed_price'
x = a[['mileage','year','Accident','Past Owners','is_Fleet_Use','is_Personal_Use']]
y = a[outcome_name]

## Perform the regression
model = LinearRegression().fit(x, y)
```

Plot the prediction result together with true price value

```python
y_predict=model.predict(x)
plt.plot([x for x in range(len(y_predict))],y,color='green',linewidth=1.0,linestyle='--')
plt.plot([x for x in range(len(y_predict))],y_predict,color='red',linewidth=1.0,linestyle='--')
```
